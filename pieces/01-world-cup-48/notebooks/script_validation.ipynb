```python
import pandas as pd
df = pd.read_csv('../data/processed/elo_at_wc_entry.csv')

# Germany 2014 and Brazil 1970
checks = df[
    ((df['team'] == 'Germany') & (df['year'] == 2014)) |
    ((df['team'] == 'Brazil') & (df['year'] == 1970))
][['year', 'team', 'elo_at_entry']]

print(checks.to_string(index=False))

# Also print top 5 highest rated teams ever at WC entry
print('\nTop 10 highest Elo at WC entry:')
print(df.nlargest(10, 'elo_at_entry')[['year', 'team', 'elo_at_entry']].to_string(index=False))
```
